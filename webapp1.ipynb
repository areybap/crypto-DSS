```python
%%writefile app.py
import streamlit as st
import pickle
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense


st.set_option('deprecation.showfileUploaderEncoding', False)
st.set_option('deprecation.showPyplotGlobalUse', False)
st.set_option('deprecation.showPyplotGlobalUse', False)

# Define the paths to your pretrained model files
model1_path = '/content/drive/MyDrive/satwick/trained_models/dt.pkl'
model2_path = '/content/drive/MyDrive/satwick/trained_models/lr.pkl'
model3_path = '/content/drive/MyDrive/satwick/trained_models/rf.pkl'
model4_path = '/content/drive/MyDrive/satwick/trained_models/svm.pkl'
model5_path = '/content/drive/MyDrive/satwick/trained_models/xgb.pkl'
model6_path = '/content/drive/MyDrive/satwick/trained_models/lda.pkl'
model7_path = '/content/drive/MyDrive/satwick/trained_models/nb.pkl'
model8_path = '/content/drive/MyDrive/satwick/trained_models/lstmnew.pkl'



# Load the models from .pkl files
with open(model1_path, 'rb') as f:
    model1 = pickle.load(f)

with open(model2_path, 'rb') as f:
    model2 = pickle.load(f)

with open(model3_path, 'rb') as f:
    model3 = pickle.load(f)

with open(model4_path, 'rb') as f:
    model4 = pickle.load(f)

with open(model5_path, 'rb') as f:
    model5 = pickle.load(f)

with open(model6_path, 'rb') as f:
    model6 = pickle.load(f)

with open(model7_path, 'rb') as f:
    model7 = pickle.load(f)

with open(model8_path, 'rb') as f:
    model8 = pickle.load(f)


# Define the prediction functions for each model
def predict_model1(input_data):
    # Perform preprocessing and prediction using model1
    prediction = model1.predict(input_data)
    return prediction

def predict_model2(input_data):
    # Perform preprocessing and prediction using model2
    prediction = model2.predict(input_data)
    return prediction

def predict_model3(input_data):
    # Perform preprocessing and prediction using model2
    prediction = model3.predict(input_data)
    return prediction

def predict_model4(input_data):
    # Perform preprocessing and prediction using model2
    prediction = model4.predict(input_data)
    return prediction

def predict_model5(input_data):
    # Perform preprocessing and prediction using model2
    prediction = model5.predict(input_data)
    return prediction

def predict_model6(input_data):
    # Perform preprocessing and prediction using model2
    prediction = model6.predict(input_data)
    return prediction

def predict_model7(input_data):
    # Perform preprocessing and prediction using model2
    prediction = model7.predict(input_data)
    return prediction

def predict_model8(input_data):
    # Perform preprocessing and prediction using model2
    prediction = model8.predict(input_data)
    return prediction



# Define the Streamlit web app
def main():
    st.title("Crypto Prediction App")

    # Add a sidebar for user input
    st.sidebar.title("User Input")

    input1 = st.sidebar.text_input("open", '43046.58')
    input2 = st.sidebar.text_input("high", '43046.58')
    input3 = st.sidebar.text_input("low", '43046.58')
    input4 = st.sidebar.text_input("volume", '0.142977')

    # Concatenate the user inputs into a NumPy array
    user_input = np.array([input1, input2, input3, input4], dtype=float).reshape(1,-1)

    lstm_input = np.array([[input1], [input2], [input3], [input4]], dtype=float).reshape(1, 1, 4)


    # Add a dropdown to select the model
    model_selection = st.sidebar.selectbox("Select Model", ["Decision Tree", "Logistic Regression", "Random Forest", "Support Vector Machine", "X Gradient Boost", "Hierarchial Bayesian Model", "Naive Bayes", "Long Short Term Memory"])

    # Perform prediction based on the selected model and user input
    if model_selection == "Decision Tree":
        # Call the prediction function for model 1 with the user input
        prediction = predict_model1(user_input)
        st.subheader('You have selected :blue[_Decision Tree_] :sunglasses:')
        #st.metric(label="Accuracy", value="0.755177803829621")
        st.write("Accuracy ", 0.5339976553341149)
        st.write("Precision: ", 0.53)
        st.write("Recall: ",  0.53 )
        st.write("F1-score: ",  0.53)
        if prediction == 0:
            st.header('You should  :red[_SELL_]')
        else:
            st.header('You should :green[_BUY_]')

    elif model_selection == "Logistic Regression":
        # Call the prediction function for model 2 with the user input
        st.subheader('You have selected :blue[_Logistic Regression_] :sunglasses:')
        st.write("Accuracy ", 0.755177803829621)
        st.write("Precision: ", 0.76)
        st.write("Recall: ",  0.75 )
        st.write("F1-score: ",  0.75 )
        prediction = predict_model2(user_input)
        if prediction == 0:
            st.header('You should  :red[_SELL_]')
        else:
            st.header('You should :green[_BUY_]')

    elif model_selection == "Random Forest":
        st.subheader('You have selected :blue[_Random Forest_] :sunglasses:')
        st.write("Accuracy ", 0.5427901524032825)
        st.write("Precision: ", 0.56)
        st.write("Recall: ",  0.55 )
        st.write("F1-score: ",  0.52 )
        prediction = predict_model3(user_input)
        if prediction == 0:
            st.header('You should  :red[_SELL_]')
        else:
            st.header('You should :green[_BUY_]')

    elif model_selection == "Support Vector Machine":
        st.subheader('You have selected :blue[_SVM_] :sunglasses:')
        st.write("Accuracy ", 0.5168034388432982)
        st.write("Precision: ", 0.26)
        st.write("Recall: ",  0.50 )
        st.write("F1-score: ",  0.34 )
        prediction = predict_model4(user_input)
        if prediction == 0:
            st.header('You should  :red[_SELL_]')
        else:
            st.header('You should :green[_BUY_]')

    elif model_selection == "X Gradient Boost":
        st.subheader('You have selected :blue[_XGB_] :sunglasses:')
        st.write("Accuracy ", 0.755177803829621)
        st.write("Precision: ", 0.76)
        st.write("Recall: ",  0.75 )
        st.write("F1-score: ",  0.75 )
        prediction = predict_model5(user_input)
        if prediction == 0:
            st.header('You should  :red[_SELL_]')
        else:
            st.header('You should :green[_BUY_]')

    elif model_selection == "Hierarchial Bayesian Model":
        st.subheader('You have selected :blue[_Bayesian model_] :sunglasses:')
        st.write("Accuracy ",  0.7532239155920282)
        st.write("Precision: ", 0.76)
        st.write("Recall: ",  0.75 )
        st.write("F1-score: ",  0.75 )
        prediction = predict_model6(user_input)
        if prediction == 0:
            st.header('You should  :red[_SELL_]')
        else:
            st.header('You should :green[_BUY_]')

    elif model_selection == "Naive Bayes":

        prediction = predict_model7(user_input)
        st.subheader('You have selected :blue[_Naive Bayes_] :sunglasses:')
        st.write("Accuracy ", 0.4957014458772958)
        st.write("Precision: ", 0.51)
        st.write("Recall: ",  0.50 )
        st.write("F1-score: ",  0.50 )
        if prediction == 0:
            st.header('You should  :red[_SELL_]')
        else:
            st.header('You should :green[_BUY_]')

    elif model_selection == "Long Short Term Memory":
        st.subheader('You have selected :blue[_LSTM_] :sunglasses:')
        st.write("Accuracy ", 0.90572958)
        st.write("RMSE: ", 0.92)


        prediction = predict_model8(lstm_input)
        st.header('The closing price should be:')
        st.header(prediction[0][0])

    else:
        prediction = "Please select a model"


    # Display the prediction result
    #st.write("Prediction:", prediction)

    import pandas as pd

    # submit button
    clickSubmit = st.button('save')
    d = {'open': input1,
            'high': input2,
            'low': input3,
            'volume': input4,
            'Prediction': prediction}
    df = pd.DataFrame(data=d)

    from csv import writer

    if clickSubmit == True:
        st.markdown('<h5>saves!</h5>', unsafe_allow_html=True)

        st.markdown('Submitted responses:')
        st.write(df)

        with open('df.csv', 'a') as f_object:
            writer_object = writer(f_object)

            writer_object.writerow(d.values())
            f_object.close()

    else:
        st.markdown("Click save to save the results")


if __name__ == "__main__":
    main()
```
